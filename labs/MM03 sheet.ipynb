# Question 1

```python
def get_curve(n):
    num = 6 * 100
    x = np.linspace(-3, 3, num)
    y = np.sin(2*np.pi*n*x)
    return y

def get_sin(n=1):
    x = np.linspace(-3, 3, 600)
    y = np.sin(n*x)
    return y
```

```python
x = np.linspace(-3, 3, 6 * 100)
y1 = get_curve(0.25)
y2 = get_curve(1)
y3 = get_curve(2)

plt.plot(x, y1, label='n=0.25')
plt.plot(x, y2, label='n=1')
plt.plot(x, y3, label='n=2')
# plt.plot(x, (y1 + y2 + y3), label='Sum')
plt.legend()
```
